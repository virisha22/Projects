```python
import sys
import os
import pandas as pd
import dask.dataframe as dd
from gatspy.periodic import LombScargleFast
df = pd.read_csv('asu.tsv', sep='\t', header=0, engine = 'python')
grouped = df.groupby(df.Source)
df.head()
```

```python
def createSeries(x):
    ind = x['ObsTime']
    y = x['Gmag']
    y.index = ind
    return y
lightcurves = grouped.apply(createSeries)
inds = lightcurves.index
inds.levels[0]
temp =[]
for i in range(0,3194):
        obj_id = inds.levels[0][i]
        sys.stdout = open(os.devnull,"w") #To prevent printing intermediate results by in-built functions
        ls= LombScargleFast(); #For finding periodicity in irregularly sampled data
        ls.optimizer.period_range = (0.1, 1.0);
        ls.fit(lightcurves[obj_id].index,lightcurves[obj_id]);
        period = ls.best_period;
        sys.stdout = sys.__stdout__
        temp.append(period)
print temp 

    
```

```python
df1 = pd.DataFrame({'Period': temp})
df1['Source']=inds.levels[0][:3194]
df1.to_csv('a.tsv', sep='\t')
df1
```
