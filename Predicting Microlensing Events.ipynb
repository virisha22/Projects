```python
import numpy as np
import pandas as pd
df = pd.read_csv('result.csv', header=0) ##Reading a pre-processed csv file
df

```

```python
#Proper motion indicates the astronomical measure of change in position of the star over time

df['pmra2']= df.pmra*df.pmra ##Calculating the square of 'Proper Motion in Right Ascension' direction
df['pmdec2']= df.pmdec*df.pmdec ##Calculating the square of 'Proper Motion in Declination' direction
df['pm2']= df.pmra2 + df.pmdec2 
df['pmmas']= df['pm2'].map(np.sqrt) ##Calculating the 'Proper Motion'
myint=1000
df['pmarc']= df.pmmas/myint ##Converting ito mas/yr
mycond = 0.15 ##threshold for high proper motion 
df['microlensing event']= df.pmarc >= 0.15
df['microlensing event'].value_counts() ##counting the number of microlensing events
```
